```python
from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
consumer = KafkaConsumer(
    'stock-project-kafka1',
     bootstrap_servers=['98.85.251.250:9092'], #add your IP here
    value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
# for c in consumer:
#     print(c.value)
```

```python
s3 = S3FileSystem()
```

```python
for count, i in enumerate(consumer):
    with s3.open("s3://aman-kafka-stock-market-project/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value, file)
```
